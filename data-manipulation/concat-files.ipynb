```python
import pandas as pd
```

```python
def concat_files(category: str, years: list[int], months: list[int]) -> pd.DataFrame:
    df = pd.DataFrame()
    for y in years:
        for m in months:
            path = f'../data/raw/{category}{y:02d}{m:02d}.csv'
            print(f'Concatenating Volumetry={df.size} : Path={path}')
            df = pd.concat([df, pd.read_csv(path, sep=';', dtype=str)])
    return df
```

```python
df = concat_files('RJRS', range(8, 25), range(1, 13))
df.to_csv('../data/RJRS_concat.csv')

df = concat_files('ERRS', range(11, 25), range(1, 13))
df.to_csv('../data/ERRS_concat.csv')

df = concat_files('RDRS', range(8, 25), range(1, 13))
df.to_csv('../data/RDRS_concat.csv')
```
